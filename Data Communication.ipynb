```python
# unipolar
import numpy as np
import matplotlib.pyplot as plt

signal = input()
signal_voltage_levels = list()
for pulse in signal:
    if(pulse=='0'):
        signal_voltage_levels.append(0)
    else:
        signal_voltage_levels.append(5)
signal_voltage_levels.append(0)

time = np.arange(0,len(signal_voltage_levels))

plt.ylabel("unipolar")
plt.step(time,signal_voltage_levels,where='post')
plt.show()
    
```

```python
# Manchester
import numpy as np
import matplotlib.pyplot as plt

signal = input()

signal_voltage_levels = []

for pulse in signal:
    if(pulse=='0'):
        signal_voltage_levels.append(1)
        signal_voltage_levels.append(-1)
    else:
        signal_voltage_levels.append(-1)
        signal_voltage_levels.append(1)

time = np.arange(0,len(signal_voltage_levels))

plt.step(time,signal_voltage_levels,where='post')
plt.ylabel("Manchester")
plt.show()
```

```python
# Differential Manchester
import numpy as np
import matplotlib.pyplot as plt

signal = input()

signal_voltage_levels = []

i = [-1,1]
j = [1,-1]
prev = i
for pulse in signal:
    if(pulse=='0'):
        signal_voltage_levels.append(prev[0])
        signal_voltage_levels.append(prev[1])
    else:
        if(prev == i):
            prev = j
        else:
            prev = i
        signal_voltage_levels.append(prev[0])
        signal_voltage_levels.append(prev[1])

time = np.arange(0,len(signal_voltage_levels))
plt.step(time,signal_voltage_levels,where='post')
plt.ylabel("Differential Manchester")
plt.show()
```

```python
# Amplititude Shift Keying

import numpy as np
import matplotlib.pyplot as plt

signal = input()

zero_amplititude = 0
one_amplititude = 5
carrier_frequency = 1
phase = 0
start = 0

modulated_signal = np.array([])
modulated_time = np.array([])

for bit in signal:
    time = np.arange(start,start+2,0.001)
    
    if(bit=='0'):
        # A.sin(omega.t + del) or #A.sin(2.pie.f.t + del)
        temporary_signal = zero_amplititude*np.sin(2*np.pi*carrier_frequency*time + phase)
    else:
        temporary_signal = one_amplititude*np.sin(2*np.pi*carrier_frequency*time + phase)
    modulated_signal = np.append(modulated_signal,temporary_signal)
    modulated_time = np.append(modulated_time,time)
    start = start+2
        
plt.plot(modulated_time,modulated_signal)
plt.ylabel('Amplititude Shift Keying')
plt.show()
```

```python
# Frequency Shift Keying

import numpy as np
import matplotlib.pyplot as plt

signal = input()

amp = 5
zero_frequency = 1
one_frequency = 3
phase = 0
start = 0

m_signal = np.array([])
m_time = np.array([])

for bit in signal:
    time = np.arange(start,start+2,.001)
    if(bit == '0'):
        tmp_signal = amp*np.sin(2*np.pi*zero_frequency*time+phase)
    else:
        tmp_signal = amp*np.sin(2*np.pi*one_frequency*time+phase)
    
    start = start+2
    m_signal = np.append(m_signal,tmp_signal)
    m_time = np.append(m_time,time)

    
plt.ylabel('Frequency Shift Keying')
plt.plot(m_time,m_signal)
plt.show()
```

```python
# Phase Shift Keying

import numpy as np
import matplotlib.pyplot as plt

signal = input()

start = 0
amp = 5
freq = 2
phase = 0

mSignal = np.array([])
mTime = np.array([])

for bit in signal:
    time = np.arange(start,start+2,.001)
    if(bit=='0'):
        tmpSignal = amp*np.sin(2*np.pi*freq*time+phase)
    else:
        tmpSignal = amp*np.sin(2*np.pi*freq*time*(-1)+phase)
    
    start = start+2
    mSignal = np.append(mSignal,tmpSignal)
    mTime = np.append(mTime,time)
    
plt.ylabel('Phase Shift Keying')
plt.plot(mTime,mSignal)
plt.show()
```

```python
# Amplititude Modulation

import numpy as np
import matplotlib.pyplot as plt

ampM = 3
ampC = 7
freqM = 5
freqC = 20
time = np.arange(0,1,.001)
messegeSignal = ampM*((np.sin(2*np.pi*freqM*time))**3 - (np.cos(2*np.pi*freqM*time)**2))
carrierSignal = ampC*np.sin(2*np.pi*freqC*time)
modulatedSignal = (ampC+messegeSignal)*np.cos(2*np.pi*freqC*time)

plt.ylabel('Amplititude Modulation')

plt.subplot(3,1,1)
plt.title('Messege Signal')
plt.plot(time,messegeSignal)

plt.subplot(3,1,2)
plt.title('Carrier Signal')
plt.plot(time,carrierSignal)

plt.subplot(3,1,3)
plt.title('Modulated Signal Signal')
plt.plot(time,modulatedSignal)

plt.subplots_adjust(hspace=1)
fig = plt.gcf()
fig.set_size_inches(8,9)
plt.show()
```

```python
# Frequency Modulation

import numpy as np
import matplotlib.pyplot as plt

ampM = 3
ampC = 7
freqM = 5
freqC = 30
time = np.arange(0,1,.001)
messegeSignal = ampM*((np.sin(2*np.pi*freqM*time)))
convertedMSignal = messegeSignal.astype(int)
carrierSignal = ampC*np.sin(2*np.pi*freqC*time)

for value in convertedMSignal:
    if(convertedMSignal[value])>=0:
        phase = .5*convertedMSignal
    else:
        phase = -0.5*(convertedMSignal)

modulatedSignal = ampC*np.sin(2*np.pi*freqC*time+phase)


plt.ylabel('Amplititude Modulation')

plt.subplot(3,1,1)
plt.title('Messege Signal')
plt.plot(time,messegeSignal)

plt.subplot(3,1,2)
plt.title('Carrier Signal')
plt.plot(time,carrierSignal)

plt.subplot(3,1,3)
plt.title('Modulated Signal Signal')
plt.plot(time,modulatedSignal)

plt.subplots_adjust(hspace=1)
fig = plt.gcf()
fig.set_size_inches(8,9)
plt.show()
```
